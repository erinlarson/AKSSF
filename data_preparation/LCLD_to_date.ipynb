# Convert lcld to Julian Date
lcld - last day of the longest continuous snow season
Spanning two years - year of and year prior dates recorded as day number

```python

import arcpy, os, sys
arcpy.env.overwriteOutput = True
lcldin = r"C:\\Users\\dwmerrigan\\Documents\\GitHub\\ArcGIS_Default\\ArcGIS_Default.gdb\\LCLD2012RasterToPoly_AddDates"

rastname = os.path.basename(lcldin)
rastyear = 2012
#rastyear = int(rastname.split("_")[0])
#lcldin = arcpy.TableToTable_conversion(lcldin,r'memory','lcldin')
arcpy.AddField_management(lcldin,'year',"SHORT")
arcpy.AddField_management(lcldin,'doy_conv',"SHORT")
arcpy.AddField_management(lcldin,'date',"DATE")
print(rastname)
print(rastyear)
print(lcldin)
for f in arcpy.ListFields(lcldin):
    print(f.name)

fields = ['gridcode','year','doy_conv','date']
from datetime import datetime, date, timedelta
with arcpy.da.UpdateCursor(lcldin,fields) as cur:
    #tablename = os.path.basename(lcldin.getOutput(0))
    #print(tablename)
    for row in cur:
        print (f'\n{("*")*100}')
        # adjusting day num
        doy = row[0] - 365
        year = rastyear
        # Initializing start date
        strt_date = date(year, 1, 1)
        # converting to date
        res_date = strt_date + timedelta(days=doy - 1)
        res = res_date.strftime("%m-%d-%Y")
        # printing result
        print(f"LCLD Value:{row[0]} - 365 = {doy} (Resolved date: {str(res)})")
        row[1] = year
        row[2] = doy
        row[3] = res
        print (f'{("*")*100}\n')
        cur.updateRow(row)
    del(row)
del(cur)

# with arcpy.da.SearchCursor(lcldin,'Value') as cur:
#     tablename = os.path.basename(lcldin.getOutput(0))
#     print(tablename)
#     for row in cur:
#         print (f'{row[0]}\n{("*")*100}')
#         if row[0] <= 366:
#             # adjusting day num
#             doy = row[0]
#             year = rastyear - 1
#             # Initializing start date
#             strt_date = date(year, 1, 1)
#             # converting to date
#             res_date = strt_date + timedelta(days=doy - 1)
#             res = res_date.strftime("%m-%d-%Y")
#             # printing result
#             print(f"LCLD {doy} Resolved date :  {str(res)}")
#             # row[1] = year
#             # row[2] = doy
#             # row[3 = res
#         else:
#             # adjusting day num
#             doy = row[0]-365
#             year = rastyear
#             # Initializing start date
#             strt_date = date(year, 1, 1)
#             # converting to date
#             res_date = strt_date + timedelta(days=doy - 1)
#             res = res_date.strftime("%m-%d-%Y")
#             print(f"Resolved date :  {str(res)}")
#             # row[1] = year
#             # row[2] = doy
#             # row[3 = res
#         print (f'{("*")*100}\n')
```
