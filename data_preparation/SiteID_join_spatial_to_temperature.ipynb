### Populate new site ID field using relational Dictionary


```python
# Fields to be queried and updated
fields = ['Site', 'SiteID_temp']
# Lists to store records that match/missing from idDict
missing = []
matches = []

in_dat = r"C:\\Users\\dwmerrigan\\Documents\\GitHub\\ArcGIS_Default\\ArcGIS_Default.gdb\\sites_wTScovars"
c=0
b=0
with arcpy.da.UpdateCursor(in_dat,fields) as cur:
    for row in cur:
        key = row[0].lower()
        #print(key)
        if key in idDict:
            #print(f'{c}. {row[0]} will be changed to {idDict[key]}')
            matches.append(row[0])
            row[1] = idDict[key]
            cur.updateRow(row)
            c+=1
        else:
            #print(f'{b}. {row[0]} not in idDict and will not be updated')
            missing.append(row[0])
            b+=1
print (f'Total updates = {c}')
print (f'Total unrelated records = {b}')
print(f'{missing}')
```
